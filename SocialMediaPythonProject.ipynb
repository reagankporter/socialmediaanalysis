**Importing Packages**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Cleaning First CSV File**

```python
file = "smmh.csv"
smmh_df = pd.read_csv(file)
```

```python
def update_word(word):
    if isinstance(word, str):
        word_lower = word.lower()  # Convert to lowercase
        if word_lower.startswith('f'):
            return 'F'
        elif word_lower.startswith('m'):
            return 'M'
    return word

# Apply the function to the specified column
smmh_df['Gender'] = smmh_df['Gender'].apply(update_word)

smmh_df = smmh_df.dropna()

int_columns = ['Age', 'Often using social media without a specific purpose?', 'On a scale of 1 to 5, how easily distracted are you?', 'On a scale of 1 to 5, how much are you bothered by worries?', 'Do you find it difficult to concentrate on things?', 'On a scale of 1-5, how often do you compare yourself to other successful people through the use of social media?', 'Following the previous question, how do you feel about these comparisons, generally speaking?', 'How often do you look to seek validation from features of social media?', 'How often do you feel depressed or down?', 'On a scale of 1 to 5, how frequently does your interest in daily activities fluctuate?', 'On a scale of 1 to 5, how often do you face issues regarding sleep?']

for col in int_columns:
    smmh_df[col] = smmh_df[col].replace({'\$': '', ',': '', '\.': ''}, regex=True)
    smmh_df[col] = pd.to_numeric(smmh_df[col], errors='coerce').astype(int)  # Convert to integers and handle NaN values
```

**Cleaning Second CSV File**

```python
file = "smts.csv"
smts_df = pd.read_csv(file)
```

```python
def update_word2(word):
    if isinstance(word, str):
        word_lower = word.lower()  # Convert to lowercase
        if word_lower.startswith('female'):
            return 'F'
        elif word_lower.startswith('male'):
            return 'M'
        elif word_lower.startswith('non-binary'):
            return 'NB'
    return word

# Apply the function to the specified column
smts_df['Gender'] = smts_df['Gender'].apply(update_word2)

smts_df = smts_df.dropna()

int_columns = ['Age', 'Time spent?', 'Income']

for col in int_columns:
    smts_df[col] = pd.to_numeric(smts_df[col], errors='coerce').astype(int)  # Convert to integers and handle NaN values
```

```python
smts_df
```

**Merging Created Dataframes**

```python
merged_df = pd.merge(smmh_df, smts_df, how='inner')
```

**Which age groups spend the most time on Social Media?**

```python
# Calculate average time spent on social media for each age group
average_time_by_age = smts_df.groupby('Age')['Time spent?'].mean()

# Plot the data
plt.plot(average_time_by_age.index, average_time_by_age.values)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Time')
plt.title('Average Time Spent on Social Media by Age')

# Display the plot
plt.show()
```

```python
# Calculate average time spent on social media for each age group
average_time_by_age = smts_df.groupby('Age')['Time spent?'].mean()

# Plot the data
plt.bar(average_time_by_age.index, average_time_by_age.values)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Time')
plt.title('Average Time Spent on Social Media by Age')

# Display the plot
plt.show()
```
